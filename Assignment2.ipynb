## SET B

### Q1. Set values of column between 0 and 1

```python
from sklearn.datasets import fetch_california_housing
```

```python
cali_housing = fetch_california_housing(as_frame=True)
cali_housing_df = cali_housing.frame
cali_housing_df.head()
cols = cali_housing_df.columns
print(cols)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
cali_housing_df = scaler.fit_transform(cali_housing_df)
cali_housing_df = pd.DataFrame(cali_housing_df, columns = cols)
cali_housing_df.head()
```

### Q2. Split dataset into training and test

```python
from sklearn.model_selection import train_test_split
```

```python
x = cali_housing_df.iloc[:,:-1]
x
y = cali_housing_df.iloc[:, -1]
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.20)
print(x_train, x_test, y_train, y_test)
```
